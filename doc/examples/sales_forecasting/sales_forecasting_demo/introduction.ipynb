# Sales Forecasting with Exasol

In this notebook we are performing predictions of sales on future dates based on the sales data of
past dates of a retail store. To perform this we will be using a dataset of a retail store that we have
included in our .zip file. Alrenatively, you can also choose to use your own dataset. 
The purpose of this demo is to showcase the AI-architecture of Exasol and not to implement the best practices to train forecasting models.

The machine learning model that, we will be using in our demo is [XGBoost](https://www.nvidia.com/en-us/glossary/xgboost/).

## Prerequisites

Before going any further. The following step need to be completed.
1. Visit the [main_config](../main_config.ipynb) notebook and complete all the steps to set up your **prefered database**.

**Note: This tutorial currently does not support a SaaS backend.**
